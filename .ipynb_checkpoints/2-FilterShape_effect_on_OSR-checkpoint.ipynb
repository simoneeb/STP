# Functions

```python
#periodic flashes 
def periodic_flashes (frequency, flashnumber,stimulus_length=stimulus_length,delay=delay,flashduration=flashduration,polarity = polarity, dt = dt, baseline = 0) :

    period = 1/frequency
    #length =int( (period * flashnumber + 2 * delay)/dt) 
    stimulus = np.zeros(int(stimulus_length/dt))
    tp_startflashes = int(delay/dt)
    tp_endflashes = tp_startflashes + int((flashnumber * period)/dt) - int(1)

    stimulus[0:tp_startflashes] = baseline

    i=-1
    for tp in range (tp_startflashes, tp_endflashes) :
        i=i+1
        if i% (period/dt) < (flashduration/dt):
            stimulus[tp] = polarity
        else:
            stimulus[tp] = baseline

    stimulus[tp_endflashes:] = baseline

    return stimulus,tp_startflashes,tp_endflashes



def impulse_train(stimulus_length,frequency, flashnumber,delay,flashduration,amplitude = polarity, dt = dt, baseline = 0):
    stim = np.zeros(int(delay/dt))
    for fn in range(flashnumber):
        stim = np.append(stim,unit_impulse(int((1/frequency)/dt))*amplitude)
        
    stim = np.append(stim,np.zeros(int(stimulus_length/dt)-len(stim)))
    tp_imps = time[(np.where(np.diff(stim) == -1)[0])]
    return [stim, tp_imps]


# model to predict, 4 params
def linear_filter_model(t,params):
    
    tau1 = params[0]
    tau2 = params[1]
    amp1 = params[2]
    amp2 = -1*params[3]
    t0 = params[4]
    
    res = np.zeros(len(t))
    
    for i,ti in enumerate(t): 
 
        if ti < t0:
            res[i] =  0
        else: 
            res[i] = amp1 * ((ti-t0)/tau1) * np.exp(1 - (ti-t0)/tau1)  + amp2 * ((ti-t0)/tau2) * np.exp(1 - (ti-t0)/tau2) 
    
    return res

   
def slope(x,s,i):
    return s * x + i

def N(x,slope =1, threshold = 0):
    if x <= threshold:
        return 0
    else:
        return slope*(x-threshold)

    


```

## Case 2 : Period smaller than filter duration but bigger than half the filter duration


Still no delay shift !!



The rebound response to flash n overlaps with the exitatory response to flash n+1 and is thus smaller than the filter bump. The more overlap, the smaller the response. When the period is half the finter length, there will be no response at all. The last bump has no overlap and is thus similar for all frequencies, the delay between the last peak and the last flash is again exactly equal. 



UPDATE: this is only the case when the two bumps are symmetric themselves. If they are not, there will be a small resonse to flashes until the peaks of rebound from flash n and excitation from flash n+1 line up, that is when the period is equal to the distance betwen the two bumps in the filter (??? CHECK). From then onwards there is no response.

## Case 3 : Period smaller than  half the filter duration

When the bump to the last flash starts, the response from the previous flash has not yet ended, resulting in a summation of both responses. The higher the frequencies, the more overlap, resulting in a smaller delay between the last peak and the last flash for higher frequencies.
