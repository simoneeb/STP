```python
# compute error
def compute_mse(y,prediction):
    
    T = len(y)
    error = 0
    for i,yi in enumerate(y): 
        error += (prediction[i]-yi)**2
    error /= T
    error = np.sqrt(error)
    return error





def optimize_evol(func,x0,nb_repeats = 200, popsize = 20, sigma0 = 0.03):
    x0 = np.log(x0)
    es = cma.CMAEvolutionStrategy(x0, sigma0, {'popsize' : popsize})

    for n in range(nb_repeats):
        
        paramsets = es.ask()
        errors = []
        for set in paramsets:
            pred = func(filtertime,np.exp(set))
            err = compute_mse(time_rf_smooth,pred)
            errors.append(err)

        es.tell(paramsets,errors)
        es.disp()

    return es

```

# Fit Linear fiter of the model, biphasic as difference of alpha profiles

```python
# model to predict, 4 params
def linear_filter_model(t,params):
    
    tau1 = params[0]
    tau2 = params[1]
    amp1 = params[2]
    amp2 = -params[3]
    t0 = params[4]
    
    res = np.zeros(len(t))
    
    for i,ti in enumerate(t): 
 
        if ti < t0:
            res[i] =  0
        else: 
            res[i] = amp1 * ((ti-t0)/tau1) * np.exp(1 - (ti-t0)/tau1)  + amp2 * ((ti-t0)/tau2) * np.exp(1 - (ti-t0)/tau2) 
    
    return res


time = np.arange(0,1,.02)


```

# Make predictions with LN model 

Take the filter fitted to STA and convolve with stimulus, then pass it through a piecewise linear nonlinearity. 

```python
#load stimuli and cell responses 
experiment_name = '20210326'
cell = 242

#load cell response to euler 

def get_euler_response(cell, experiment_name):
    filename = f"/user/sebert/home/Documents/Experiments/Results/{experiment_name}_Thomas/resultfiles/{experiment_name}_euler_grade_isi.pkl"
    with open(filename, "rb") as handle:   #Pickling
        infodict = pickle.load(handle)
    
    return infodict[f'temp_{cell}']['control']['euler_count']


#load  euler stim stimulus 
def get_euler_stimulus():
    stimeuler = pd.read_csv('/user/sebert/home/Documents/Experiments/StimulusDesgin/Euler_Baptiste/euler_luminance_profile.csv')
    euler = np.asarray(stimeuler['luminance'])
    euler = euler/euler.mean()
    euler = euler - euler.std()
    euler_time = np.arange(0,30,0.02)
    
    return euler,euler_time



#load osr info
def get_OSR_response(cell,experiment_name, IDs):
    
    filename = f"/user/sebert/home/Documents/Experiments/Results/{experiment_name}_Thomas/resultfiles/osr_raster_count_peak_slope.pkl"
    with open(filename, "rb") as handle:   #Pickling
        osrdict = pickle.load(handle)
        
    key = f'temp_{cell}'
    
    responses = []
    osr_times = []
    flashstarts = []
    flashends = []
    lastflashends = []
    peaks = []
    
    for ID in IDs:
        
        #get responses
        responses.append(osrdict[key]['control'][ID]["rate_smoothed"])

        #get plotting stuff
        
        flashstarts.append(osrdict['stimulus_info'][ID]['flashstart'])
        flashends.append(osrdict['stimulus_info'][ID]['flashend'])
        osr_times.append(osrdict['stimulus_info'][ID]["osr_time"])
        lastflashends.append(osrdict['stimulus_info'][ID]["lastflashend"])
        peaks.append(osrdict[key]['control'][ID]["peak"])
    return responses,osr_times,flashstarts,flashends,lastflashends,peaks
        

#OSR Stim
def get_OSR_stimulus(experiment_name):
    result_filepath = f'/user/sebert/home/Documents/Experiments/Results/{experiment_name}_Thomas/resultfiles'
    plotstimulus = pd.read_csv(os.path.join(result_filepath, 'plot_stimulus.csv'), index_col = 0)
    IDs = np.array(plotstimulus.columns)
    stimulus_info = pd.read_csv(f'/user/sebert/home/Documents/Experiments/StimulusDesgin/Stimulus_JAN21_OSR_AmacrineInhibition_50Hz/stimulus_OSR_jan21ID_info.csv')
    return plotstimulus, stimulus_info, IDs
```

```python
# predictions


def design_matrix_1D(stimulus,td=40):
    
    stimulus_timesteps = stimulus.shape[0]+td
    
    
    stimulus_padded = np.concatenate((np.zeros(td),stimulus))
    X = np.zeros((stimulus_timesteps,td))
    
    for t in range(stimulus_timesteps):
        if t < td:
            continue
        else:
            X[t,:] = stimulus_padded[t-td:t]
    
        
    return X[td+1:,:]
    
 

    
def convolve(stim,sta, dimension = '1D'):
    
    
    if dimension == '2D':
        X = design_matrix_2D(stim)
        sta = sta.ravel()
        
    if dimension == '1D':
        X = design_matrix_1D(stim)
  
    response = np.concatenate((np.zeros(1),np.dot(X,sta)))
    return  response




def N(l,slope = 1 ,threshold = 0):
    
    if l <= threshold:
        return 0
    else:
        return l*slope

    
    
# temporal LN model with two pathways and piecewise nonlinear rectification

def predict(stim,params,slope = 1, threshold = 0):
   
    
    filt = linear_filter_model(filtertime,params)
    
    linear_prediction = convolve(stim,np.flip(filt))

    nonlinear_prediction = np.array([N(l,slope =slope,threshold = threshold) for l in linear_prediction])
    
    
    return linear_prediction, nonlinear_prediction
    

def slope(x,s,i):
    return s * x + i
```

# Predict OSR response

```python
stimuli_df,info,IDs = get_OSR_stimulus(experiment_name)
osr_responses,osr_times,flashstarts,flashends,lastflashends,peaks = get_OSR_response(c,experiment_name,IDs)
delays = np.array(peaks) - np.array(lastflashends)


periods = info['period'].unique()*1000
info
```

```python

stimuli= []
predictions_linear = []
predictions_nonlinear = []
delays_predictions = []
for i,ID in enumerate(IDs):
    stimulus = stimuli_df[ID].dropna().values

    stimulus -= stimulus[0]
    stimulus /= np.abs(stimulus).max()

    #make linear prediction
    model_linear_response, model_nonlinear_response = predict(stimulus,bestparams,slope = 5, threshold = 0)
    
    predictions_linear.append(model_linear_response)  
    predictions_nonlinear.append(model_nonlinear_response)
    delays_predictions.append(osr_times[i][np.argmax(model_nonlinear_response)]-lastflashends[i])
    stimuli.append(stimulus)
```
